```
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```
img1 = cv2.imread('/content/football.jpg',0)
plt.imshow(img1,cmap = 'gray')
```

```
img2 = cv2.rotate(img1, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img2, cmap = 'gray')
```

```
imginfloat = np.float32(img1)
img3 = cv2.cornerHarris(imginfloat,1,31,0.04)
plt.imshow(img3)
```

```
imginfloat = np.float32(img2)
img3 = cv2.cornerHarris(imginfloat,1,31,0.04)
plt.imshow(img3)
```

```
img = cv2.imread('football.jpg',0)
img_r = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
# Initiate FAST object with default values
fast = cv2.FastFeatureDetector_create() # <- FastFeatureDetector()

# find and draw the keypoints
kp = fast.detect(img,None)
img1 = cv2.drawKeypoints(img, kp, np.array([]), color=(255,0,0))
plt.imshow(img1)
```

```
kp = fast.detect(img_r,None)
img2 = cv2.drawKeypoints(img_r, kp, np.array([]), color=(255,0,0))
plt.imshow(img2)
```

```
img = cv2.imread('football.jpg',0) # queryImage
img_r = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
orb = cv2.ORB_create()        # Initiate SIFT detector
```

```
# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(img, None)
img1 = cv2.drawKeypoints(img, kp1, np.array([]), color=(255,0,0))
plt.imshow(img1)
```

```
# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(img_r, None)
img2 = cv2.drawKeypoints(img_r, kp1, np.array([]), color=(255,0,0))
plt.imshow(img2)
```
